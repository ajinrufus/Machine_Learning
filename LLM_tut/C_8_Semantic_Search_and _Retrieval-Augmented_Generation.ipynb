# Semantic Search

It allows search by context/meaning and not only by keywords. eg: used by google and bing nowadays.

Categories of how to Use language model for search:
1. Dense retrieval
   * Works similar to embedding model. Converts search and all documents to a embedding and gives nearest neigbours as search result.



   * The first step is to create embedding for the documents by using chunks of vector per document or use single vector per document.
   * The above limitation is because transformers can operate only upto a certain context length.
   * Therefore for single vector, it needs to take only certain parts of the document or aggregate the chunks to a single vector. But, much of the information is lost.



  * Different chunking ways:



2. Reranking
   * Use two steps:
     * retrieve results based on query.
     * use a rerank LM to rank the documents according to their relevance with the query.



* each intial result is checked with the query for relevance and ranked according to their relevance score.
* check monoBERT for detail - https://arxiv.org/abs/1910.14424

3. Retrieval Augmented Generation (RAG)

This has gained advantage with the arrival of LLMs that the user expects answers rather than some links. But since the answers can be wrong sometimes creating hallucinations, it is better sometimes to provide answers in a way such as below. this is what RAG does.



It uses the documents generated in first stage as a context and answers the query. The seconjd LLM is called grounded LLM because the documents generated acts a context and grounds the LLM to the domain we are interested in.



RAG evaluation metric is still ongoing... https://arxiv.org/abs/2304.09848 is one method for reading.
