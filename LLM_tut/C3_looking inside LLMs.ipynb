# Transformer model for text generation (decoder only)



# 1. Transformer block



## a. Attention layer:
* Used to get the relevant information (context) from all other input tokens. It adds information into the representation (embedding vector) of the last token to be used for the succeeding text generation.
* To give the transformer more attention capability, the attention mechanism is duplicated multiple times in parallel (attention head).



* `Steps`
1. projection
   
   

2. scoring each token association with currect token (last token)

   

3. combining information by giving attention to each value

   


## b. feed forward layer:


Choosing the highest probability is called `greedy decoding`. Best way is to perform sampling of probability distribution.

# Recent Advancements to transformer model

## 1. More Efficient Attention
### a. Sparse attention

This can limit the performance, therefore not used throughout. GPT-3 uses it in alternate blocks.





each row tells how many previous tokens it attends.

### b. Grouped query and shared query



* Each embedding vector shares its key and value vectors.



* As the model size grows, this can be punishing. Thats why grouped qwuery attention is used.



## 2. Transformer block

### a. Normalisation imporvements:
* Normalisation before attention and MLP is found to be faster.
* Using RMSNorm instead of layerNorm seems to be more efficient.

### b. Activation function
* SwiGLU instead of ReLU is more common nowadays.



## 3. Packaging
* Sequences vary in length. To have same length (for GPU processing), they are padded. But it is found that almost 50% of sequence is padding, Therefore, without padding, the training speed can be increased 2X.



* Simple concatenating (packaging strategy) or sing [SEP] token can contaminate the attention and therefore can affect the performance (reduction in F1 score continuously).

* Algorithm that can be used should,
    * Efficient algorithm to decide which smaples to put together to have as little padding as possible.
    * Alter BERT model to process the pack w/o contamination
    
https://www.graphcore.ai/posts/introducing-packed-bert-for-2x-faster-training-in-natural-language-processing

## 4. RoPe

* Rotary positional embedding is one of the most common method to prevent contamination.
* They are applied in the attention step to the query and key matrices just before we multiply them and not at the start of the forward pass.
